```python
payload = {
            "account[username]": "abcd@gmail.com", # your email address
            "account[password]": "abcdefghi", # your password
}
```

```python
import requests
from lxml import html
```

```python
session_requests = requests.session()
```

```python
login_url = "https://www.gdinside.com/login.php"
result = session_requests.get(login_url)

# tree = html.fromstring(result.text)
```

```python
result = session_requests.post(
    login_url, 
    data = payload, 
    headers = dict(referer=login_url)
)
```

```python
result.text
```

```python
url = 'https://www.gdinside.com/category.php?id=1&view_all=1'
result = session_requests.get(
    url, 
    headers = dict(referer = url)
)
```
